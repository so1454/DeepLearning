### 목표

```python
# 데이터 불러오기
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
# csv (이 파일의 구분자가 ';' 임)
data = pd.read_csv('./data/student-mat.csv',delimiter = ';')
data
```

```python
#  컬럼명만 확인
data.columns

# 답 데이터 : G3 (수학 성적)
# 특성 데이터 : 특성 1개만 학습 예정 => studytime(공부할 시간)
```

```python
# 문제와 답으로 분리
X = data.loc[ : , 'studytime'] # 공부한 시간 1차원
y = data.loc[ : , 'G3'] # g3 1차원 - 회귀

print(X.shape)
print(y.shape)

# train, test 분리
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state= 3)

# 크기 확인
print('훈련용',X_train.shape)
print('훈련용',y_train.shape)
print('테스트용',X_test.shape)
print('테스트용',y_test.shape)
```

### 신경망 모델 만들기
- 1. 신경망 구조 설계 : 뼈대 구축, 층 쌓기
- 2. 신경망 학습 / 평가하는 방법 설정
- 3. 학습 및 학습 결과 시각화
- 4. 모델 평가 및 예측

```python
# 1. 신경망 구조 설계
# 뼈대 구축 도구 불러오기
from tensorflow.keras import Sequential
# 층 쌓기
from tensorflow.keras.layers import Dense, Activation

# Dense : 층 내용 정의하는 기능
# Activation : 활성화 함수

model = Sequential() # 뼈대 구축

# 층 쌓기 => 입력층, 중간층(은닉층), 출력층
# 유닛 == 뉴런
# 입력층 + 중간층 첫번째(뉴런(유닛)개수가 4개로 구성!)
# model.add(Dense(중간부분의 유닛개수, 입력층의 입력데이터 개수(크기)설정 => 특성 열의 개수만큼))
# model.add(Activation(특정 함수)) # 이렇게 두개 해줘야 마무리!
              # 활성화 함수 : 반응 결과 정리하는 기능

model.add(Dense(units = 4, input_dim = 1 ))
model.add(Activation('sigmoid'))  

# 출력층
# 성적 예측 => 회귀
# '''회귀'''할때 항상 units =1 ! , 활성화 함수 : 'linear'(y=wx+b라는 뜻) 를 항상 사용한다! linear는 생략이 가능! => 항등함수
model.add(Dense(units = 1))

# 요약
model.summary()



```

```python
# 2. 신경망 학습/ 평가 방법 설정
# compile()
# 모델의 설계를 pc가 이해할 수 있도록 번역하는 과정
# 오차 계산 공식, 최적화 함수, 평가 지표
model.compile(loss = 'mse', #손실함수(비용함수) : mse(평균제곱 오차) => 회귀일때 사용
              optimizer = 'sgd' # 최적화 함수 : 경사하강법 종류 설정 => 확률적 경사하강법
              ) # metrics = ['accuracy'] metrics는 mse로 활용(생략함)
```

```python
# 학습
# h라는 변수에 학습된 현황 저장
h = model.fit(X_train,y_train, epochs = 200)
h

# 반복하면서 성능이 점점 좋아지고있음 (loss가 점점 낮아짐)
```

```python
# 모델 평가
# mse 활용 => 오차 출력
model.evaluate(X_test,y_test)
```

```python
# 모델 예측
# predict()
model.predict(X_test)  # 공부시간에 따라서 수학 성적 예측해주라에 대한 예측 결과임
```
